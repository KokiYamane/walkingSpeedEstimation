```
# 相補フィルター
def complementary_filter(df_acc):
    C = 0.99
    angle_list = [0.0]
    for t in df_acc.index[1:]:
        dt = 0.02
        gyro = angle_list[-1] + df_acc['gyroX[deg/s]'][t] * dt
        mag = df_acc['angleZ[deg]'][t]
        angle = C * gyro + (1-C) * mag
        angle %= 360
        angle_list.append(angle)
    angle_list = np.array(angle_list)
    return angle_list

angle_list = complementary_filter(df_acc)

# num = 50 # 移動平均の個数
# b = np.ones(num)/num
# angle_list = np.convolve(angle_list, b, mode='same') # 移動平均
```

```
# 初期位置と初期速度を代入した「4次元状態」
x = np.array([df_acc['angleZ[deg]'][0], df_acc['gyroX[deg/s]'][0]]) 

# 共分散行列
P = np.array([[0.1, 0.],
              [0., 0.1]])

# 状態遷移行列
dt = 0.02 # 計測間隔
F = np.array([[1., dt],
              [0., 1.]])

# 観測行列
H = np.array([[1., 0.],
              [0., 1.]])

# ノイズ
R = np.array([[0.0, 0.0],
              [0.0, 0.0]])

# 4次元単位行列
I = np.identity(2)


def kalman_filter(x, P):
    x_list = []
    P_list = []

    for t in df_acc.index:
        # 予測
        x = ld.deglim(np.dot(F, x))
        P = np.dot(np.dot(F, P), F.T)

        # カルマンゲイン
        S = np.dot(np.dot(H, P), H.T) + R
        K = np.dot(np.dot(P, H.T), np.linalg.inv(S))

        # 計測更新
        Z = np.array([df_acc['angleZ[deg]'][t], df_acc['gyroX[deg/s]'][t]])
        y = Z.T - np.dot(H, x)
        x = x + np.dot(K, y)
        P = np.dot((I - np.dot(K, H)), P)

        x_list.append(x)
        P_list.append(P)

    return np.array(x_list), np.array(P_list)

x_list, P_list = kalman_filter(x, P)
print(x_list)
print(P_list[:, 0, 0])
```

```
def step_estimation(df_acc):
    accX = df_acc['accX[m/s^2]']
    accY = df_acc['accY[m/s^2]']
    accZ = df_acc['accZ[m/s^2]']
    acc_norm = np.sqrt(accX**2 + accY**2 + accZ**2)
    acc_norm = acc_norm.rolling(5).mean()
    acc_norm = acc_norm / 9.8 - 1

    step_list = []
    acc_old = acc_norm[0]
    for acc in acc_norm:
        step = 0
        g = 0.2
        if acc_old < g and acc > g: step = 1
        acc_old = acc
        step_list.append(step)
    
    return step_list

step_list = step_estimation(df_acc)

plt.plot(acc_norm.as_matrix(), color='tab:blue')
plt.plot(step_list, color='tab:green')
start = 10000
plt.xlim(start, start+200)
```

```
import datetime

def distance_estimation(df_acc):
    step_estimation(df_acc)
    distance_list = [0.0]
    for t in tqdm(df_acc.index):
        if df_acc['step'][t] == 0:
            distance_list.append(distance_list[-1])
            continue

        start = t - datetime.timedelta(seconds=0.5)
        acc = df_acc['accX[m/s^2]'][start:t]
        k = 0.5
        stride = k * (max(acc) - min(acc))**(1/4)
        distance_list.append(distance_list[-1] + stride)
    
    return distance_list

distance_list = distance_estimation(df_acc)
```
